### Data Description

first_name - firstname of the person.  
middle_initial  - middle name of the person.  
last_name - last name of the person.   
suffix - suffix 
system - system  
agency -agency 
organization - Departments.  
subtitle - subtitle  
class_code - class code.  
annual_salary - annual salary of a a person  
pay_rate - Grades for the payrate.  
regular_earnings - Actual salary of a person.  
overtime_earnings - Salary for overtime working hours.  
other_earnings - other earnings offered by the company like bonus etc.  
ytd_gross_earnings - gross earnings of a person  
term_date - term date of person in series

```python
#visulation of organization using histogram
plt.figure(figsize=(16,12))
plt.hist(data['organization'])
plt.xticks(rotation=90)
plt.xlabel("Organization",fontsize=12)
plt.ylabel("Frequency",fontsize=12)
plt.title("Organization",fontsize=14)
plt.show()
```

```python
#visualisingthe relationship b/w numerical variables 
plt.figure(figsize=(16,12))
plt.plot(data['organization'],data['annual_salary'])
plt.xticks(rotation=90)
plt.show()
```

```python
#visualising the regular earning using histogram
plt.figure(figsize=(14,10))
plt.hist(data['regular_earnings'])
plt.xticks(rotation=90)
plt.xlabel("Regular Earnings",fontsize=10)
plt.ylabel("Frequency",fontsize=10)
plt.title("Earnings",fontsize=14)
plt.show()
```

```python
#visualising the relation between organization and regular eranings
plt.figure(figsize=(16,12))
plt.plot(data['organization'],data['regular_earnings'])
plt.xticks(rotation=90)
plt.show()
```

```python
#visualisation of regular ernings using distribution plot
plt.figure(figsize=(6,6))
sns.distplot(data['regular_earnings'])
plt.show()
```

```python
#Converting categorical variables into Numerics
from sklearn.preprocessing import LabelEncoder
categorical_features=['organization','pay_rate']
encoder=LabelEncoder()
data[categorical_features]=data[categorical_features].apply(encoder.fit_transform)
```
